```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Let's pull the candidates info 

house_cand = pd.read_excel("house.xlsx")
print(house_cand.columns)
house_cand.head()
```

```python
sns.set()
in_state_winners = house_cand[(house_cand["GE WINNER INDICATOR"] == 1) & (house_cand["STATE ABBREVIATION"] == house_cand["STATE ABBREVIATION"])].count()[0]
in_state_losers = house_cand[(house_cand["GE WINNER INDICATOR"] == 0) & (house_cand["STATE ABBREVIATION"] == house_cand["STATE ABBREVIATION"])].count()[0]

out_state_winners = house_cand[(house_cand["GE WINNER INDICATOR"] == 1) & (house_cand["STATE ABBREVIATION"] != house_cand["STATE ABBREVIATION"])].count()[0]
out_state_losers = house_cand[(house_cand["GE WINNER INDICATOR"] == 0) & (house_cand["STATE ABBREVIATION"] != house_cand["STATE ABBREVIATION"])].count()[0]

in_state_plot = pd.DataFrame({"result":['Win','Lose'],"No of Candidates":[in_state_winners,in_state_losers]})
out_state_plot = pd.DataFrame({"result":['Win','Lose'],"No of Candidates":[out_state_winners,out_state_losers]})
g = sns.factorplot(x='result', y='No of Candidates',data=in_state_plot, kind='bar',size=7, aspect=0.5)
f = sns.factorplot(x='result', y='No of Candidates',data=out_state_plot, kind='bar',size=7, aspect=0.5)

for ax in g.axes.flatten():
    ax.set_xlabel('')
    ax.set_title('In-State candidate')
for ax in f.axes.flatten():
    ax.set_xlabel('')
    ax.set_title('Out-State candidate')
    
g.savefig("in-state.jpg",dpi=1000)
f.savefig("out-state.jpg",dpi=1000)
```
