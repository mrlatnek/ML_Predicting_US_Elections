```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Datasets to merge
datasets = ["final_data/new_2012.csv","final_data/new_2014.csv","final_data/new_2016.csv"]
no_of_datasets = len(datasets)

def concat_datasets(datasets):
    ''' 
    Create one dataframe out of all these datasets.
    It should have State, District, Party, Winner
    Return a dataframe
    '''
    data = [pd.read_csv(i) for i in datasets]
    dfs = [df.filter(items=['Cand_Office_St','Cand_Office_Dist', 'Cand_Party_Affiliation',
        'GE WINNER INDICATOR']) for df in data]
    main_df = pd.concat(dfs)
    
    # Change Nan to 0 and W to 1 in winning cand column
    main_df['GE WINNER INDICATOR'] = main_df['GE WINNER INDICATOR'].fillna(0)
    main_df['GE WINNER INDICATOR'] = main_df['GE WINNER INDICATOR'].replace("W",1)
    return main_df
```

```python
main_df = concat_datasets(datasets)
# main_df.head()
```
