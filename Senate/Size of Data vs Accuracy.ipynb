```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import roc_auc_score
```

```python
# Read 
senate = pd.read_csv("senate_analytical.csv")

# Randomize the data
senate = shuffle(senate)

senate.head()
```

```python
# Reading the test data
senate_test = pd.read_csv("test_analytical.csv")

# print(senate_test["Cand_Office_St"].value_counts())

y_test = senate_test["GE WINNER INDICATOR"]

test_states = senate_test["Cand_Office_St"]
X_test = senate_test.drop(columns=["Cand_Office_St","Cand_Office_Dist","Cand_Id","Cand_Incumbent_Challenger_Open_Seat","Cand_Name","Cand_Office","Cand_State","GE WINNER INDICATOR","Individual_Contribution","Other_Committee_Contribution","Party_Committee_Contribution","Total_Disbursement","Total_Receipt"])


print(X_test.shape)
X_test.columns

# y_test = y_test.astype(int)
X_test.head()
```

### For 1 year data (2016)

```python
senate_2016 = senate[senate["year"] == 2016].copy()

y = senate_2016["GE WINNER INDICATOR"]
X = senate_2016.drop(columns=["Cand_Id","Cand_Incumbent_Challenger_Open_Seat","Cand_Name","Cand_Office","Cand_State","GE WINNER INDICATOR","Individual_Contribution","Other_Committee_Contribution","Party_Committee_Contribution","Total_Disbursement","Total_Receipt","year"])

print(y.value_counts())
print(X.shape)
# y.head()

senate_2016.head()
```

```python
clf = RandomForestClassifier(n_estimators=200)
clf.fit(X,y)
X_test_1 = X_test.copy()
y_prob_1 = clf.predict_proba(X_test_1)[:,1]
print("Score: " + str(clf.score(X_test_1,y_test)))
```

```python
X_test_1["State"] = test_states
X_test_1["y_prob"] = y_prob_1

X_test_1["winner"] = 0
X_test_1.loc[X_test_1.groupby(['State'])["y_prob"].idxmax(),"winner"] = 1

print("AUC: " + str(roc_auc_score(y_test, y_prob_1)))
precision_recall_fscore_support(y_test, X_test_1["winner"])
```
